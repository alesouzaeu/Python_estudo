# Importando dados (IPCA - Banco Central)

```
import pandas as pd
import seaborn as sns;
import matplotlib.pyplot as plt
%matplotlib inline
```

## Consumindo API do Banco Central

```
def consulta_bc(codigo_bcb):
      url = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.433/dados?formato=json'.format(codigo_bcb)
      df = pd.read_json(url)
      df['data'] = pd.to_datetime(df['data'], dayfirst=True)
      df.set_index('data', inplace=True)
      return df
```

```
ipca = consulta_bc(433)
ipca
```

## Plotando dataframe

```
ipca.plot(kind="line", figsize=(20,10))
```
